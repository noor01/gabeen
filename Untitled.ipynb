```python
from controllers import fluid_control
import networkx as nx
```

```python
fluid_controller = fluid_control(['blah'],'squidward','MERFISH_20_bit')
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
color_map = {
    'Valve': 'red',
    'Pump': 'blue',
    'source': 'green',
    'destination': 'purple'
}
```

```python
G = fluid_controller.graph
```

```python
G.nodes
```

```python
node_colors = [color_map[G.nodes[node]['node_type']] for node in G.nodes]
```

```python
nx.draw(G, with_labels=True, node_color=node_colors, node_size=800)
plt.show()
```

```python
start_node = '4'
end_node = 'waste'

# Find the shortest path considering weights
path_nodes = nx.shortest_path(G, source=start_node, target=end_node, weight='weight')

# Extract edges from the path
path_edges = list(zip(path_nodes[:-1], path_nodes[1:]))

print(path_edges)
```
